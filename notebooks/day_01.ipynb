```python
file_path = '../data/day_01.txt'

with open(file_path, 'r') as file:
    lines = file.readlines()
    lines = [line.strip() for line in lines]
```

```python
numeric_digits = []

for line in lines:
    digits = ''.join(filter(str.isdigit, line))
    if digits:
        numeric_digits.append(digits)

numeric_digits
```

```python
calibration_values = []
for number in numeric_digits:
    if len(number) == 1:
        calibration_values.append(int(number * 2))
    elif len(number) > 2:
        number = number[0] + number[-1]
        calibration_values.append(int(number))
    else:
        calibration_values.append(int(number))

sum_calibration_values = sum(digit for digit in calibration_values)
calibration_values
sum_calibration_values
```

```python
results = []

for line in lines:
    temp = []
    i = 0

    while i < len(line):
        if line[i].isdigit():
            temp.append(line[i])
            i += 1
        else:
            for c in str_digits:
                if c in line[i:i + len(c)]:
                    temp.append(str(str_digits.index(c)))
                    i += 1
                    break
            else:
                i += 1

    results.append(temp)

results
```

```python
calculations = []
for line in results:
    first = line[0]
    last = line[-1]
    total = int(first + last)

    calculations.append(total)

calculations = sum(calculations)
calculations
```
